```python
import pandas as pd
import numpy as np
import requests
import json
from nyt_config import api_key
from pprint import pprint
```

```python
base_url = "https://api.nytimes.com/svc/books/v3"
tail = f"?api-key={api_key}"
```

```python
response = requests.get(f"{base_url}/lists/names.json{tail}")
data=response.json()
```

```python
pprint(data)
```

```python
data['results'][24]['list_name']
```

```python
result_count = np.arange(data['num_results'])
```

```python
lists_list=[]
new_pub=[]
old_pub=[]
```

```python
for i in result_count:
    list_name = (data['results'][i]['list_name'])
    lists_list.append(list_name)
    new_pub_date = (data['results'][i]['newest_published_date'])
    new_pub.append(new_pub_date)
    old_pub_date = (data['results'][i]['oldest_published_date'])
    old_pub.append(old_pub_date)
```

```python
lists_df = pd.DataFrame({
    "List": lists_list,
    "Newest_Publication_Date": new_pub,
    "Oldest_Publication_Date": old_pub
})
```

```python
lists_df=lists_df.sort_values(by='Newest_Publication_Date', ascending=False)
```
